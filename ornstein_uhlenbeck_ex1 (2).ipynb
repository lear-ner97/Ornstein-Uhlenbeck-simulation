```python

import numpy as np
import matplotlib.pyplot as plt

num_sims = 1000

t_init = 0#
t_end  = 1#
N      = 100  # Compute 100 grid points
dt     = float(t_end - t_init) / N
x_init = 1

lambdaa = 0.5#
q=1.0


def dB(delta_t):
    """Sample a random number at each call."""
    return np.random.normal(loc=0.0, scale=np.sqrt(q*delta_t))#loc=mean,scale=std



#vector of time steps
ts = np.arange(t_init, t_end + dt, dt)#start,stop,step
#vector of x values
xs = np.zeros(N + 1)
xs[0] = x_init

mean=0
variance=np.zeros(num_sims)
for j in range(num_sims):
    for i in range(1, ts.size):
        t = t_init + (i - 1) * dt
        x = xs[i - 1]
        xs[i] = x - lambdaa * x * dt +  dB(dt)
    mean+=xs[-1]
    variance[j]=xs[-1]
    plt.plot(ts, xs)

plt.xlabel("time (s)")
h = plt.ylabel("x")
h.set_rotation(0)
plt.show()
print('mean= '+str(mean/num_sims))
print('variance= '+str(np.var(variance)))
```

THE computated meand and variance are close to their theoretical values
